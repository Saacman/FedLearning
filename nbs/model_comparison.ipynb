```python
import torch, os
from collections import OrderedDict
from torchvision import models
from torchsummary import summary
from fedlern.models.resnet_v2 import ResNet18
model_dir = "../saved_models"
# Define the list of state dicts to load
state_dicts_path = ['resnet_fedlern_dynbits_2.pth']

# Load the state dicts
state_dicts = []
for state_dict_path in state_dicts_path:
    path = os.path.join(model_dir, state_dict_path)
    state_dict = torch.load(path)
    state_dicts.append(state_dict)

# Get the keys (parameter names) from the first state dict
parameter_names = list(state_dicts[0].keys())[:5]

# Define the column width for each parameter
column_width = 20

# Print the parameter names in the first row
for parameter_name in parameter_names:
    print(f"{parameter_name:{column_width}}", end="")
print()

# Print the values for each parameter in separate columns
for state_dict, name in zip(state_dicts, state_dicts_path):
    print(name)
    for parameter_name in parameter_names:
        parameter_value = state_dict[parameter_name]
        print(parameter_value)
        #print(f"{parameter_value:{column_width}.4f}", end="")
    print()
```

```python
model = ResNet18()
for state_dict, name in zip(state_dicts, state_dicts_path):
    print(name)
    model.load_state_dict(state_dict)
    model.cuda()
    summary(model, input_size=(3, 32, 32))
```
